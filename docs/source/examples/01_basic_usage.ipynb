# Basic Usage of `missingly`

This notebook provides a basic walkthrough of the main features of the `missingly` package.

## 1. Creating a sample dataset

```python
import pandas as pd
import numpy as np
import missingly as mi

data = {
    'A': [1, 2, np.nan, 4, 5, np.nan],
    'B': [10.0, np.nan, 30.0, 40.0, 50.0, 60.0],
    'C': [np.nan, np.nan, np.nan, np.nan, 100, 110]
}
df = pd.DataFrame(data)
df
```

## 2. Summary Functions

```python
mi.miss_var_summary(df)
```

## 3. Visualizations

```python
mi.matrix(df)
```

```python
mi.dendrogram(df)
```
